## Exercise 1
This program will a file, break each line into words, strip whitespace and punctuation from the words, and convert them to lowercase.
Input: any .txt file or the default file: test.txt.
Output: Output_Ex1.txt will be generated in the same folder (ensheng_stat504/HW1 in this case).

```python
import string # For removing punctuations
import os.path # Check whether a file exists or not

print "\n\nStat 504 hw 1 Exercise 1"
print "This program will a file, break each line into words, strip whitespace"
print "and punctuation from the words, and convert them to lowercase."
print "Output_Ex1.txt will be generated in the same folder as the input file.\n"

# Get the sequence of words from the file
fileRead = raw_input("Use the default test.txt? (Y/N): ") or "Y"   
if fileRead == "Y" or fileRead == "y":
    fname = 'test.txt'
elif fileRead == "N" or fileRead == "n":
    fname = raw_input("File to analyze: ")
    if fname == "" or not(os.path.isfile(fname)):
        fname = raw_input("Wrong input! We will use the default file.\nPress Enter to continue.")
        fname = 'test.txt'
else:
    fname = raw_input("\nWrong input! We will use the default file.\nPress Enter to continue.")
    fname = 'test.txt'
fin = open(fname)
print ("\nYour input file: %s\n\nResults:\n" % fname)

# Output file
text_file = open("Output_Ex1.txt", "w")

# Break lines into words
lines = fin.readlines()

for line in lines:
    # Replace punctuation with space
    replace_punctuation = string.maketrans(string.punctuation, ' '*len(string.punctuation))
    line = line.translate(replace_punctuation)
    #print "\n" + line
    words = line.split()
    for word in words:
        # Remove punctuations in each break
        # word = word.translate(None, string.punctuation)
        # Print the list of words in lowercase
        text_file.write(word.lower() + '\n')
        print (word.lower())

text_file.close()
print "\nPlease check the output file in Output_Ex1.txt!"
```
